```python
#1 Check if a number is palindrome or not
n = int(input())
rev = 0
temp = n
while(n!=0):
  d = n%10
  rev = rev*10 + d
  n = n//10
if(rev == temp):
  print("Palindrome")
else:
  print("no")
```

```python
#2 Find all Palindrome numbers in a given range
min = int(input())
max = int(input())
for n in range(min,max+1):
  temp = n
  rev = 0
  num = n
  while(num!=0):
    d = num%10
    rev = rev*10 + d
    num = num//10
  if temp == rev:
    print(temp, end = " ")

```

```python
#22 two prime numbers sum is n
n = int(input())
def prime(n):
  if n<=1:
    return False
  for i in range(2,int(n**0.5)+1):
    if n%i == 0:
      return False
  return True

for i in range(2,n):
  if prime(i) and prime(n-i):
    print(i,n-i)
    break
else:
  print("no")
```

```python
#27Program to Add two fractions
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

num1, den1 = 1, 3
num2, den2 = 1, 6

lcm = (den1 * den2) // gcd(den1, den2)

n1 = num1 * (lcm // den1)
n2 = num2 * (lcm // den2)

num = n1 + n2
den = lcm

g = gcd(num, den)

num //= g
den //= g

print(num, "/", den)
```
